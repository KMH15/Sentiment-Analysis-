```python
import pandas as pd
import nltk
nltk.download('punkt')
import re
!pip install emoji
!pip install -U emoji
import emoji
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from gensim.models import Word2Vec
from nltk.corpus import brown
```

```python
df=pd.read_csv("labeled_data.csv")
df=df[:10000]
print(df.head())
print(df.info())

```

```python
df_words=[]
for i in range(len(df['tweet'])):
  df_words.append(str(df['tweet'][i]))
token=[]
for i in range(len(df_words)):
  token.append(nltk.word_tokenize(df_words[i]))
print(token[i])

```

```python
def cleaning(inp_text):
  text = ' '.join(inp_text)
  text = re.sub(r'@[A-Za-z0-9]+', '', text)
  text = re.sub(r'#[A-Za-z0-9]+', '', text)
  text = re.sub(r'https?://\S+', '', text)
  text = re.sub(r'www\.[^ ]+', '', text)
  text = re.sub(r'[^\w\s]', '', text)
  text = emoji.demojize(text)
  text = text.lower()
  words = nltk.word_tokenize(text)
  stop_words = set(stopwords.words('english'))
  words = [word for word in words if word not in stop_words]
  lemmatizer = WordNetLemmatizer()
  words = [lemmatizer.lemmatize(word) for word in words]
  text = ' '.join(words)
  print(text)
  return text
```

```python
cleaned_text=[]
for i in range(len(token)):
  cleaned_text.append(cleaning(token[i]))
print(cleaned_text[0])
#print(cleaned_text)
```

```python
X_train, X_test, y_train, y_test = train_test_split(cleaned_text,df['class'],test_size=0.2, random_state=42)
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train_vec=vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)



```

```python
clf = SVC(kernel='linear')
clf.fit(X_train_vec, y_train)
y_pred = clf.predict(X_test_vec)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
dc = DecisionTreeClassifier(max_depth=10)

dc.fit(X_train_vec, y_train)

y_pred_2 = dc.predict(X_test_vec)

print(classification_report(y_test, y_pred_2))
print("Accuracy:", accuracy_score(y_test, y_pred_2))
```

```python
LRmodel = LogisticRegression(C = 2, max_iter = 1000, n_jobs=-1)
LRmodel.fit(X_train_vec, y_train)
y_pred1 = LRmodel.predict(X_test_vec)
print(classification_report(y_test, y_pred1))
print("Accuracy:", accuracy_score(y_test, y_pred1))
```
